```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
As = 0.1923e-9
ns = 0.965
k_pivot = 0.05                                                            # Mpc^-1
k_check = np.logspace(-4, 1, 1000)                                       # Rango de k (logspace para cubrir varios rdenes de magnitud)

# Ley de potencias
Pk_base = As*(k_check/k_pivot)**(ns-1.0)                                  # P(k) = As * (k / k_pivot)^(ns - 1)

# Oscilaciones Suaves
A_osc = 0.05                                                               # A = Amplitud de la oscilacin (ej. 0.2 para que sea visible pero no gigante)
omega = 0.1                                                              # omega = Frecuencia de la oscilacin
factor_oscilacion = 1 + A_osc * np.sin(omega *(k_check / k_pivot))
#factor_oscilacion = 1 + A_osc * np.sin(omega * np.log(k_check / k_pivot))

# Nueva potencias 
Pk_camb = Pk_base * factor_oscilacion

# Grfica 
plt.figure(figsize=(10, 6))

plt.plot(k_check, Pk_camb, color='blue', lw=2, label='Espectro con Oscilaciones')
plt.plot(k_check, Pk_base, color='red', linestyle='--', alpha=0.5, label='Base (Ley de Potencias)')

plt.xlabel(r'Escala $k$ [Mpc$^{-1}$]', fontsize=12)
plt.ylabel(r'Espectro de Potencia $\mathcal{P}_{\mathcal{R}}(k)$', fontsize=12)
plt.title('Espectro Primordial con Oscilaciones Suaves', fontsize=14)
plt.legend()
plt.grid(True, which="both", linestyle='--', alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.loglog(k_check, Pk_camb, color='blue', lw=2, label='Espectro con Oscilaciones')
plt.loglog(k_check, Pk_base, color='red', linestyle='--', alpha=0.5, label='Base (Ley de Potencias)')

plt.xlabel(r'Escala $k$ [Mpc$^{-1}$]', fontsize=12)
plt.ylabel(r'Espectro de Potencia $\mathcal{P}_{\mathcal{R}}(k)$', fontsize=12)
plt.title('Espectro Primordial con Oscilaciones Suaves', fontsize=14)
plt.legend()
plt.grid(True, which="both", linestyle='--', alpha=0.4)
plt.show()
```
